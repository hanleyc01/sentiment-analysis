# Graphing out authorship

Because our kmeans clustering was not the best, we hypothesize that this might be the result of the frequency of letters written by author, and that it is weighted towards a select few authors. In order to confirm this, let us graph out the distribution of letter frequencies by author.

```python
import pandas as pd
import matplotlib.pyplot as plt
import nltk
%matplotlib inline

letters = pd.read_json('letters.json')['meta-author']
```

```python
plt.figure(figsize=(20, 6))
plt_var = nltk.FreqDist(letters).plot()
```
